```
check <- 0.0
for (i in 1:1000){
  g <- erdos.renyi.game(1000, p=0.003, directed=F)
  if (is.connected(g) == 0) {
    g_not <- g
  }
  if(is.connected(g)){
    check = check + 1
  }
}
p <- check/1000
print(p)
g.components <- clusters(g_not)
g_size <- which.max(g.components$csize)
g.giant <- induced.subgraph(g, which(g.components$membership == g_size))
diameter(g.giant, directed = F)

check <- 0.0
for (i in 1:1000){
  g <- erdos.renyi.game(1000, p=0.004, directed=F)

  if (is.connected(g) == 0) {
    g_not <- g
  }
  if(is.connected(g)){
    check = check + 1
  }
}
p <- check/1000
print(p)
g.components <- clusters(g_not)
g_size <- which.max(g.components$csize)
g.giant <- induced.subgraph(g, which(g.components$membership == g_size))
diameter(g.giant, directed = F)

check <- 0.0
for (i in 1:1000){
  g <- erdos.renyi.game(1000, p=0.01, directed=F)

  if (is.connected(g) == 0) {
    g_not <- g
  }
  if(is.connected(g)){
    check = check + 1
  }
}
p <- check/1000
print(p)
g.components <- clusters(g_not)
g_size <- which.max(g.components$csize)
g.giant <- induced.subgraph(g, which(g.components$membership == g_size))
diameter(g.giant, directed = F)

check <- 0.0
for (i in 1:1000){
  g <- erdos.renyi.game(1000, p=0.05, directed=F)
  if (is.connected(g) == 0) {
    g_not <- g
  }
  if(is.connected(g)){
    check = check + 1
  }
}
p <- check/1000
print(p)
g.components <- clusters(g_not)
g_size <- which.max(g.components$csize)
g.giant <- induced.subgraph(g, which(g.components$membership == g_size))
diameter(g.giant, directed = F)

check <- 0.0
for (i in 1:1000){
  g <- erdos.renyi.game(1000, p=0.1, directed=F)
  if (is.connected(g) == 0) {
    g_not <- g
  }
  if(is.connected(g)){
    check = check + 1
  }
}
p <- check/1000
print(p)
g.components <- clusters(g_not)
g_size <- which.max(g.components$csize)
g.giant <- induced.subgraph(g, which(g.components$membership == g_size))
diameter(g.giant, directed = F)
```

```
index <- 1
frac <- integer(98)
arr<-integer(100)
for(pvalue in seq(from=0.003, to=0.1, by=0.001)){
  for (i in 1:100){
    g <- erdos.renyi.game(1000, p=pvalue, directed=F)
    g.components <- clusters(g)
    arr[i] = max(g.components$csize)
  }
  frac[index] <- mean(arr)/1000
  index<-index+1
}
x_axis<-seq(from=0.003, to=0.100, by=0.001)
plot(x_axis, frac
     , main="GCC size vs p",xlab="p value",ylab="GCC sizes",
     pch=18, cex=1)
average <- mean(frac)
abline(h = average, col = "red")
legend("bottomright", "each value", pch=18, title= "Normalized GCC 1")
legend("right", "", lwd=1, col = "red", title= "Average Normalized GCC")

# to focus on the ln(n)/n part
index_2 <- 1
frac_2 <- integer(121)
arr_2 <- integer(100)
for(p_value_2 in seq(from=0.003, to=0.015, by=0.0001)){
  for (i in 1:100){
    g <- erdos.renyi.game(1000, p=p_value_2, directed=F)
    g.components <- clusters(g)
    arr_2[i] <- max(g.components$csize)
  }
  
  frac_2[index_2] = mean(arr_2)/1000
  index_2 <- index_2+1
}
x_axis_2<-seq(from=0.003, to=0.015, by=0.0001)
plot(x_axis_2, frac_2
     , main="GCC size vs p",xlab="p value",ylab="GCC sizes",
     pch=18, cex=1)
average_2 <- mean(frac_2)
print(average_2)
abline(h = average_2, col = "red")
legend("bottomright", "each value", pch=18, title= "Normalized GCC 2")
legend("right", "", lwd=1, col = "red", title= "Average Normalized GCC")

index_3 <- 1
frac_3 <- integer(141)
arr_3 <- integer(100)
for(p_value_2 in seq(from=0.000, to=0.007, by=0.00005)){
  for (i in 1:100){
    g <- erdos.renyi.game(1000, p=p_value_2, directed=F)
    g.components <- clusters(g)
    arr_3[i] <- max(g.components$csize)
  }
  
  frac_3[index_3] = mean(arr_3)/1000
  index_3 <- index_3+1
}
x_axis_3<-seq(from=0.000, to=0.007, by=0.00005)
plot(x_axis_3, frac_3
     , main="GCC size vs p estimate emergence",xlab="p value",ylab="GCC sizes",
     pch=18, cex=1)
abline(v = 0.00095, col = "red")
legend("bottomright", "each value", pch=18, title= "Normalized GCC for Emergence")
legend("right", "", lwd=1, col = "red", title= "Estimated Emergence ")

index_3 <- 1
frac_3 <- integer(401)
arr_3 <- integer(100)
for(p_value_2 in seq(from=0.000, to=0.002, by=0.000005)){
  for (i in 1:100){
    g <- erdos.renyi.game(1000, p=p_value_2, directed=F)
    g.components <- clusters(g)
    arr_3[i] <- max(g.components$csize)
  }
  
  frac_3[index_3] = mean(arr_3)/1000
  index_3 <- index_3+1
}
x_axis_3<-seq(from=0.000, to=0.002, by=0.000005)
plot(x_axis_3, frac_3
     , main="GCC size vs p estimate emergence around start",xlab="p value",ylab="GCC sizes",
     pch=18, cex=1)
abline(v = 0.00095, col = "red")
legend("bottomright", "each value", pch=18, title= "Normalized GCC for Emergence")
legend("right", "", lwd=1, col = "red", title= "Estimated Emergence around start point")

index_3 <- 1
frac_3 <- integer(401)
arr_3 <- integer(100)
for(p_value_2 in seq(from=0.005, to=0.007, by=0.000005)){
  for (i in 1:100){
    g <- erdos.renyi.game(1000, p=p_value_2, directed=F)
    g.components <- clusters(g)
    arr_3[i] <- max(g.components$csize)
  }
  
  frac_3[index_3] = mean(arr_3)/1000
  index_3 <- index_3+1
}
x_axis_3<-seq(from=0.005, to=0.007, by=0.000005)
plot(x_axis_3, frac_3
     , main="GCC size vs p estimate emergence around 99%",xlab="p value",ylab="GCC sizes",
     pch=18, cex=1)
abline(h = 0.99, col = "red")
legend("bottomright", "each value", pch=18, title= "Normalized GCC for Emergence")
legend("right", "", lwd=1, col = "red", title= "Estimated Emergence around 99%")
```

```
frac <- matrix(integer(9901*5),nrow=9901, ncol=5)
col_index <- 1
for (candid in c(0.5, 1, 1.1, 1.2, 1.3)){
  c <- candid
  row_index<-1
  for(num_n in seq(from=100, to=10000, by=1)){
    vector <- integer(100)
    for (i in 1:100){
      g <- erdos.renyi.game(n=num_n, p=c/num_n, directed=F)
      g.components <- clusters(g)
      vector[i] = max(g.components$csize)
    }
    frac[row_index, col_index] = mean(vector)
    row_index <- row_index+1
  }
  col_index <- col_index+1
}

x_label<-seq(from=100, to=10000, by=1)
plot(x_label, frac[,1]
     , main="GCC size vs N, c=0.5",xlab="N",ylab="GCC sizes",
     ylim=range(1, 25),
     pch=18, cex=1)


plot(x_label, frac[,2]
     , main="GCC size vs n with c=1",xlab="n",ylab="GCC sizes",
     ylim=range(1, 1000),
     pch=18, cex=1)

plot(x_label, frac[,3]
     , main="GCC size vs n with c=1.1",xlab="n", col = 'yellow',ylab="GCC sizes",
     ylim=range(1, 5000),
     pch=18, cex=1)

points(x_label, frac[,4],col='red',
       pch=16, cex=0.5)
points(x_label, frac[,5],col='blue',
       pch=16, cex=0.5)
op <- par(cex = 1)
legend("left", c("c=1.1","c=1.2", "c=1.3"),
       col=c("yellow","red","blue"), pch=c(16,16,16), pt.cex=0.5, cex=0.8)
```
